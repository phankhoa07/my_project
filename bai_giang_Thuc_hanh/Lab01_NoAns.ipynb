### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.


```python
# Nhận dữ liệu đầu vào từ người dùng
x = input("Nhập ký tự x: ")
n = int(input("Nhập số vị trí n: "))

# Bước 1: Kiểm tra nếu x là chữ cái thường (a-z)
if 'a' <= x <= 'z':
    start = ord('a')  # Mốc bắt đầu là 97
    # Tính toán vị trí mới: (vị trí hiện tại - mốc + bước nhảy) chia dư 26, sau đó cộng lại mốc
    ket_qua = chr((ord(x) - start + n) % 26 + start)
    print(ket_qua)

# Bước 2: Kiểm tra nếu x là chữ cái hoa (A-Z)
elif 'A' <= x <= 'Z':
    start = ord('A')  # Mốc bắt đầu là 65
    ket_qua = chr((ord(x) - start + n) % 26 + start)
    print(ket_qua)

# Bước 3: Nếu không phải chữ cái (số, ký hiệu...), giữ nguyên ký tự đó
else:
    print(x)

```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
char_x = input()

if 'a' <= char_x <= 'z' :
   print(chr(ord(char_x)-32))
   
elif "A" <= char_x <= "Z":
   print(chr(ord(char_x)+32))
else:
   print(char_x)
   
   
```
