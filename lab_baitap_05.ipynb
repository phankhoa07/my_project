```python
def reverse_num(n):
    so_dao_nguoc = 0
    
    while n > 0:
        
        chu_so_cuoi = n % 10 
        
        
        so_dao_nguoc = (so_dao_nguoc * 10) + chu_so_cuoi 
        
        
        n = n // 10 
        
    return so_dao_nguoc
def solve_palindrome():
    try:
       
        n = int(input())
        so_buoc = 0
        
      
        while n != reverse_num(n):
            n = n + reverse_num(n)
            so_buoc += 1
            
        
        print(f"{so_buoc} {n}")
        
    except :
        print("Vui lòng nhập một số nguyên dương.")


solve_palindrome()
```

```python
a = int(input())
b = int(input())

def tong_uoc(n):
    if n <= 1:
        return 0
    tong = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            tong += i
            if i != n // i:
                tong += n // i
    return tong

if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
```
