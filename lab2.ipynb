```python
def tinh_toan(a, b):
    tong = a + b
    hieu = a - b
    phan_nguyen = a // b
    phan_du = a % b
    kq_thuc = a / b
    print(f"5 +3 =  {tong}\n 5 - 3 {hieu}\n 5 * 3 = {phan_nguyen} \n 5 % 3 = {phan_du} \n 5 / 3 = {kq_thuc} \n 5 // 3 ={phan_nguyen}")


a = int(input())
b = int(input())
tinh_toan(a , b)
```
